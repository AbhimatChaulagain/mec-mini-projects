```python
import json
import requests



api_url = 'https://data.nasdaq.com/api/v3/datasets/'
my_dataset = "FSE/AFX_X"
return_data_format = "data.json"

start_date='2017-01-01'
end_date='2017-12-31'
collapse='monthly'



```

```python
#1. Retriving only 2017 data
api_url += my_dataset + "/" + return_data_format + "?start_date=2017-01-01&end_date=2017-12-31" + "&api_key=" + API_KEY
response = requests.get(api_url)
```

```python
#2. Convering data into Dict
data_json = json.loads(response.text)
print(type(data_json))

```

```python
#3. calculating highetst and lowest opening price
open_data = []
change = []
closing_price = []
trading_volume = []
trading_volume_sum = 0
for i in data_json["dataset_data"]["data"]:
    if not i[1] is None:
        open_data.append(i[1])
    trading_volume.append((i[6]))
    trading_volume_sum += i[6]
    change.append(i[2] - i[3])
    closing_price.append(i[4])
print(f"opening Highest: {max(open_data)}, opening lowest: {min(open_data)}")

```

```python
#4. largest change in any one day based on high and low price
print(f"Largest change: {max(change)}")
```

```python
#5. Largest change between two days based on closing price
print(f"closing price: {closing_price})")
print(f"Largest change between two days based on closing price: {max(closing_price) - min(closing_price)}")
```

```python
#6 Average daily trading volume
print(f"Average trading volume: {trading_volume_sum / len(data_json['dataset_data']['data'])}")
```

```python
#7 Median trading volume
print(f"trading volume list: {trading_volume}")
sort_trading_volume = sorted(trading_volume)
print(f"trading volume list: {sort_trading_volume}")
print(f"length of sorted volume: {len(sort_trading_volume)}")
# Checking length is even and if even sum of middle two divided by 2
median = (((sort_trading_volume[len(sort_trading_volume) // 2] +
            sort_trading_volume[(len(sort_trading_volume) - 1) // 2]) / 2) if len(sort_trading_volume) % 2 == 0 else
          # If not even get middle value
          (sort_trading_volume[len(sort_trading_volume) // 2]))
print(f"Median of trading volume: {median}")
```
